##Subsetting


Subsetting allows us to filter rows based on some condition 
- Show only the "Year", "Christian", "Zoey" and "Sara" columns 
- Show only where Year is greater than 2000

```python
#Select specific columns from the dataframe

names= ["Year","Christian", "Zoey", "Sara"] 
#To select a single column, use square brackets [] with the column name of the column of interest

df[names] 
#df[["Christian", "Zoey", "Sara"]] 
```

```python
#Shows which rows have a year >2000
df.Year > 2000  
```

#####Remember, to select rows based on a conditional expression, use a condition inside the selection brackets [ ].


```python
df[ df.Year > 2000] #filters the rows where year is > 2000
```

##Creating a new Dataframe with the rows and columns we're interested in




```python
newdf = df [ df.Year > 2000]
newdf[names]
# same as 
#df[ df.index > 2000][names]
```

##Making changes to the data

####Let's create a Dataframe that has each of the names and their associated correlation with the name Karen

```python
colList = list(df.columns)
colList.remove("Year")
```

```python
#We can use a function to help us 
bigList = []
for name in colList:
    #print(df[name])
    #print(name+ ":" , df.Karen.corr(df[name]))
    bigList.append([name,df.Karen.corr(df[name])])
```

```python
karen_df = pd.DataFrame(bigList,columns=["Names", "Corr with Karen"])
karen_df.sort_values(by="Corr with Karen",ascending=False).head(26)[1:].reset_index(drop=True)
```
