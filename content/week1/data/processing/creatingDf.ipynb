```python
import pandas as pd
```

```python
df = pd.read_csv("babyTextData/yob" + "1880" + ".txt",header=None)
df["year"]= "1880"


for i in range(1881,2019):
    filename = "babyTextData/yob" + str(i) + ".txt"
    tempDf= pd.read_csv(filename, header=None)
    tempDf["year"]= str(i)
    df= pd.merge(df, tempDf,how='outer')

df.columns = ["Name", "Gender",'Count','Year']
df
```

```python
d = df.drop("Gender",axis=1)
```
