```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#using the courses tab of the Fall Application Dashboard
df = pd.read_csv("courses.csv")
```

```python
#this is gross... let's clean it
df.Gender.value_counts()
```

```python
 #take the first letter and lowercase it. Strip it of puncatation. Then deal with an edge case
df["gender_cleaned"] = df.Gender.str[0].str.lower().str.strip().str.replace("w","f")
```

```python
#helper function to make the graph
def makeGraph(group,name):
    return sns.countplot(group.gender_cleaned).set_title(name)
```

```python
#display pics inlien
%matplotlib inline

#use a count to make fig numbers
count=0

# group by course name.
#In each iteration, pull out the course name (name) and the data (group)
for name,group in df.groupby("Course"):
   #create an empty graph
    plt.figure(count)
     #add the info to graph
    makeGraph(group,name)
    count+=1
```
